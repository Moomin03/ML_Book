```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
data, target = make_moons(n_samples=200,
                          noise=0.05,
                          random_state=42)
plt.scatter(data[:, 0],
            data[:, 1])
```

```python
from sklearn.cluster import KMeans, AgglomerativeClustering
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
km = KMeans(n_clusters=2,
            random_state=42)
train_pred = km.fit_predict(data)
axes[0].scatter(data[train_pred==0, 0],
                data[train_pred==0, 1])
axes[0].scatter(data[train_pred==1, 0],
                data[train_pred==1, 1])
axes[0].scatter(km.cluster_centers_[:, 0],
                km.cluster_centers_[:, 1],
                marker='D')
ac = AgglomerativeClustering(n_clusters=2, linkage='complete')
train_pred = ac.fit_predict(data)
axes[1].scatter(data[train_pred==0, 0],
                data[train_pred==0, 1])
axes[1].scatter(data[train_pred==1, 0],
                data[train_pred==1, 1])
```

```python
from sklearn.cluster import DBSCAN
db = DBSCAN(eps=0.2)
train_pred = db.fit_predict(data)
plt.scatter(data[train_pred==0, 0],
            data[train_pred==0, 1])
plt.scatter(data[train_pred==1, 0],
            data[train_pred==1, 1])
```
