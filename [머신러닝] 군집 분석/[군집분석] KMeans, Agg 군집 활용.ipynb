```python
from sklearn.datasets import make_blobs
data, target = make_blobs(n_samples=300,
                          n_features=2,
                          centers=4,
                          cluster_std=1,
                          shuffle=True,
                          random_state=42)
import matplotlib.pyplot as plt
plt.scatter(data[:, 0],
            data[:, 1],
            color='white',
            edgecolors='black',
            linewidths=1)
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=4, init='random')
km.fit(data)
train_pred = km.predict(data)
```

```python
import numpy as np
for i in np.unique(train_pred):
    plt.scatter(data[train_pred==i, 0],
                data[train_pred==i, 1])
plt.scatter(km.cluster_centers_[:, 0],
            km.cluster_centers_[:, 1],
            marker='D', color='gray',edgecolors='black', linewidths=2)
plt.grid()
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=4, init='k-means++')
km.fit(data)
train_pred = km.predict(data)
```

```python
import numpy as np
for i in np.unique(train_pred):
    plt.scatter(data[train_pred==i, 0],
                data[train_pred==i, 1])
plt.scatter(km.cluster_centers_[:, 0],
            km.cluster_centers_[:, 1],
            marker='D', color='gray',edgecolors='black', linewidths=2)
plt.grid()
```

```python
km.inertia_
```

```python
distortions = []
for i in range(1, 11):
    km = KMeans(n_clusters=i,
                random_state=42)
    km.fit(data)
    distortions.append(km.inertia_)
plt.plot(np.arange(1, 11), distortions, marker='o')
plt.tight_layout()
```

```python
km = KMeans(n_clusters=4,
            random_state=42)
train_pred = km.fit_predict(data)
from sklearn.metrics import silhouette_samples
silhouette_vals = silhouette_samples(data, train_pred)
silhouette_vals
```

```python
from sklearn.cluster import AgglomerativeClustering
ac = AgglomerativeClustering(n_clusters=4,
                             metric='euclidean',
                             linkage='complete')
labels = ac.fit_predict(data)
print(labels)
```
