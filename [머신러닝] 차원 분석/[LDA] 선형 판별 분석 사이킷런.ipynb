```python
from sklearn.datasets import load_wine
wine_data = load_wine()
data, target = wine_data.data, wine_data.target
columns = wine_data.feature_names
```

```python
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(data,
                                                                      target,
                                                                      stratify=target,
                                                                      test_size=0.2,
                                                                      random_state=42)
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
train_scaled = ss.fit_transform(train_input)
test_scaled = ss.transform(test_input)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
lda = LDA(n_components=2)
train_lda = lda.fit_transform(train_scaled, train_target)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(multi_class='ovr', random_state=42, solver='lbfgs')
lr.fit(train_lda, train_target)
```

```python
import matplotlib.pyplot as plt
import numpy as np
x_min, x_max = train_lda[:, 0].min()-1, train_lda[:, 0].max()+1
y_min, y_max = train_lda[:, 1].min()-1, train_lda[:, 1].max()+1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                     np.arange(y_min, y_max, 0.01))
z = lr.predict(np.array([xx.ravel(), yy.ravel()]).T)
z = z.reshape(xx.shape)
plt.contourf(xx, yy, z)
colors = ['yellow', 'red', 'blue']
for color, i in zip(colors, np.unique(train_target)):
    plt.scatter(train_lda[:, 0][train_target==i], train_lda[:, 1][train_target==i], color=color)
    plt.axis('off')
```

```python
test_lda = lda.transform(test_scaled)
plt.contourf(xx, yy, z)
colors = ['yellow', 'red', 'blue']
for color, i in zip(colors, np.unique(test_target)):
    plt.scatter(test_lda[:, 0][test_target==i], test_lda[:, 1][test_target==i], color=color)
    plt.axis('off')
```
