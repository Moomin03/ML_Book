```python
from sklearn.datasets import load_breast_cancer
load_breast = load_breast_cancer()
data, columns = load_breast.data, load_breast.feature_names

import pandas as pd
breast = pd.DataFrame(data=data,
                      columns=columns)
breast.head()
```

```python
from mlxtend.plotting import scatterplotmatrix, heatmap
scatterplotmatrix(breast.iloc[:, :5].values, names=breast.iloc[:, :5].columns, figsize=(10,10))
import matplotlib.pyplot as plt
plt.tight_layout()
plt.show()
```

```python
import numpy as np
corr = np.corrcoef(breast.iloc[:, :5].values.T)
heatmap(corr, row_names=breast.iloc[:, :5].columns, column_names=breast.iloc[:, :5].columns)
```

```python
from sklearn.linear_model import RANSACRegressor, LinearRegression
ransacr = RANSACRegressor(
    estimator=LinearRegression(),
    min_samples=0.95,
    residual_threshold=None
)
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
pipe=make_pipeline(StandardScaler(),
                   ransacr)
pipe.fit(breast.iloc[:, :-1], breast.iloc[:, -1])
```

```python
pipe.score(breast.iloc[:, :-1], breast.iloc[:, -1])
```
