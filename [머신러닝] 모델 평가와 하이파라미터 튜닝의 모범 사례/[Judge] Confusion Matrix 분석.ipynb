```python
from sklearn.datasets import load_breast_cancer
breast_data = load_breast_cancer()
data, target = breast_data.data, breast_data.target
columns = breast_data.feature_names

from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(data,
                                                                      target,
                                                                      test_size=0.2,
                                                                      stratify=target,
                                                                      random_state=42)

from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
pipe = make_pipeline(StandardScaler(),
                     SVC(random_state=42))

from sklearn.metrics import confusion_matrix
pipe.fit(train_input, train_target)
test_predicted = pipe.predict(test_input)
cfm = confusion_matrix(test_target, test_predicted)
cfm
```

```python
import matplotlib.pyplot as plt
plt.matshow(cfm, alpha=0.3)
for i in range(cfm.shape[0]):
    for j in range(cfm.shape[1]):
        plt.text(x=i,
                 y=j,
                 s=cfm[i][j],
                 va='center')
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_estimator(pipe, test_input, test_target)
```

```python
ConfusionMatrixDisplay.from_estimator(pipe, test_input, test_target, normalize='all')
```

```python
from sklearn.metrics import matthews_corrcoef
matthews_corrcoef(test_target, test_predicted)
```
