```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
iris_data = load_iris()
data, target, columns = iris_data.data, iris_data.target, iris_data.feature_names
pd.DataFrame(data=data,
             columns=columns)
```

```python
# ROC AUC를 계산하기 위해서 꽃받침 너비와 꽃잎 길이 두 개의 특성만 사용하겠습니다.
# 그리고 타겟데이터의 경우 1, 2만으로 구성하겠습니다.
train_input, test_input, train_target, test_target = train_test_split(data[target!=0][:, [1, 2]],
                                                                      target[target!=0],
                                                                      random_state=42,
                                                                      test_size=0.5,
                                                                      stratify=target[target!=0])
```

```python
from sklearn.model_selection import cross_validate
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import make_pipeline
import numpy as np
model = ['LogisticRegression', 'DecisionTree', 'KNeighbors']
clf1 = LogisticRegression(penalty='l2',
                          C=0.001,
                          random_state=42)
clf2 = DecisionTreeClassifier(max_depth=1,
                              criterion='entropy',
                              random_state=42)
clf3 = KNeighborsClassifier(n_neighbors=1,
                            p=2,
                            metric="minkowski")
pipe1 = make_pipeline(StandardScaler(),
                      clf1)
pipe2 = make_pipeline(StandardScaler(),
                      clf2)
pipe3 = make_pipeline(StandardScaler(),
                      clf3)
for i, j in zip([pipe1, pipe2, pipe3], model):
    scores = cross_validate(estimator=i,
                            X=train_input,
                            y=train_target,
                            cv=10,
                            scoring='roc_auc')['test_score']
    print(f'ROC AUC : {scores.mean():.2f}' + '/' + f'+/- {scores.std():.2f}' + f' [{j}]')
```

```python
from sklearn.ensemble import VotingClassifier
vc = VotingClassifier(estimators=[('lr', pipe1), ('dt', pipe2), ('knn', pipe3)], voting='soft', verbose=True)
scores = cross_validate(estimator=vc, X=train_input, y=train_target, cv=10, scoring='roc_auc')['test_score']
print(f'ROC AUC : {scores.mean():.2f}' + '/' + f'+/- {scores.std():.2f}')
vc = vc.fit(train_input, train_target)
```

```python
# 훈련데이터에 대한 정답률 : 50개 중에서 44개 정답
(np.argmax(np.round(vc.predict_proba(test_input), 2), axis=1)==test_target-1).sum()
```

```python
from sklearn.metrics import RocCurveDisplay, roc_curve, precision_score, recall_score
import matplotlib.pyplot as plt
train_model = [pipe1, pipe2, pipe3, vc]
fig, ax = plt.subplots(figsize=(8, 5))
for i in train_model:
    i.fit(train_input, train_target)
    pred_data = i.predict_proba(test_input)[:, 1]
    fpr, tpr, thresholds = roc_curve(y_true=test_target-1,
                                     y_score=pred_data)
    RocCurveDisplay.from_estimator(i, train_input, train_target, ax=ax)
plt.grid()
plt.ylim([-0.1, 1.1])
plt.xlim([-0.1, 1.1])
```
